## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df['ConvertedComp'].plot(kind='hist')
plt.xlabel('Annual Compensation')
plt.ylabel('Number of Responders')
plt.title('Distribution of ConvertedComp of Responders')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df['Age'].plot(kind='box', figsize=(8, 6))
plt.ylabel('Age of Responders')
plt.title('Box plot of Age of Responders')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
plt.scatter(df['Age'], df['WorkWeekHrs'], color='darkblue')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter plot comparing Age and Work Week Hours')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
norm_age = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df.plot(kind='scatter',
        x='WorkWeekHrs',
        y='CodeRevHrs',
        figsize=(14, 8),
        alpha=0.5,
        color='blue',
        s=norm_age * 2000 + 10,
       )
plt.xlabel('Work Week Hours')
plt.ylabel('Code Rev Hours')
plt.title('Bubble plot comparing WorkWeekHrs and CodeRevHrs of Different Ages')
plt.legend(['Age'], loc='upper right', fontsize='xx-large')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
df = pd.read_sql_query(QUERY, conn)
df_database = df.groupby(['DatabaseDesireNextYear'], axis=0).count()
df_database.sort_values(by='Respondent', ascending=False, axis=0, inplace=True)
df_top5 = df_database.head()
df_top5['Respondent'].plot(kind='pie',
                           figsize=(7,8),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                          )
plt.title('Top 5 databases Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.legend(labels=df_top5.index, loc='upper left', fontsize=7)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)
df_grouped = df.groupby(['Age'], axis=0).median()
df_grouped.plot(kind='bar', stacked=True)
plt.ylabel('Hours')
plt.title('Comparison of Median Work Week Hours and \n Code Rev Hours for Respondents Ages 30 to 35')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)
df_grouped = df.groupby(['Age'], axis=0).median()
df_grouped.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Annual Compensation')
plt.title('Median Annual Compensation for Respondents Ages 45 to 60')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df_hbar = df['MainBranch'].value_counts()
df_hbar.plot(kind='barh', figsize=(10,5))
plt.xlabel('Number of Respondents')
plt.title('Comparison of Total Number of Respondents who are and are not a Developer')
```
